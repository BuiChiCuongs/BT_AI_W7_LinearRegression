```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
url='Advertising.csv'
data=pd.read_csv(url)
data.head()
```

```python
data.hist()
plt.show()
```

```python
sns.pairplot(data)
```

```python
correlation=data.corr()
print(correlation)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='TV',
         y='sales',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['TV']),y=data['sales'])
print('Gia_tri_intercept:',model.intercept_)
print('Gia_tri_coeffient:',model.coef_)
print('Do_chinh_xac:',model.score(X=pd.DataFrame(data['TV']),y=data['sales']))
```

```python
data.plot(kind='scatter',x='TV',y='sales',
          figsize=(9,9),color='black',xlim=(0,300))
y_pred=model.predict(X=pd.DataFrame(data['TV']))
plt.plot(data['TV'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X=pd.DataFrame([data['TV'],data['TV']**2]).T
model.fit(X,y=data['sales'])
print('Intercept:',model.intercept_)
print('Coefficient:',model.coef_)
print('Do_chinh_xac:',model.score(X,y=data['sales']))
range=np.arange(0,300,0.1)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='TV',y='sales',
          figsize=(9,9),color='black',xlim=(0,300))
plt.plot(range,y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X=pd.DataFrame([data['TV'],data['TV']**2,data['TV']**3,data['TV']**4,data['TV']**5,data['TV']**6]).T
model.fit(X,y=data['sales'])
print('Intercept:',model.intercept_)
print('Coefficient:',model.coef_)
print('Do_chinh_xac:',model.score(X,y=data['sales']))
```

```python
range=np.arange(0,300,0.1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='TV',y='sales',
          figsize=(9,9),color='black',xlim=(0,300))
plt.plot(range,y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
model.fit(X=data.loc[:,['TV','radio']],y=data['sales'])
print('Intercept:',model.intercept_)
print('Coefficient:',model.coef_)
print('Do_chinh_xac:',model.score(X=data.loc[:,['TV','radio']],y=data['sales']))
```
